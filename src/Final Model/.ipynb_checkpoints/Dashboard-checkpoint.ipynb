```python
# Scheduled Refresh of Predictions: 17:00 PM every weekday
from Functions import refreshPredictions
import schedule
import time

# set the schedule to run every weekday at 17:00 PM
schedule.every().monday.at("16:30").do(refreshPredictions)
schedule.every().tuesday.at("16:30").do(refreshPredictions)
schedule.every().wednesday.at("16:30").do(refreshPredictions)
schedule.every().thursday.at("16:30").do(refreshPredictions)
schedule.every().friday.at("16:30").do(refreshPredictions)

# run the schedule
while True:
    schedule.run_pending()
    # the schedule is once a day, so we can sleep for 1 hour
    time.sleep(3600)
    # time.sleep(3600) # 7200 seconds = 1 hour
```

```python
import plotly.graph_objects as go
    
#Get RMSE value, which is the value of RMSE column in historicalPerformance where Target = target and symbol = ticker
RMSE = .045

fig = go.Figure()   

# Add Close_True
fig.add_trace(go.Scatter(x=historicalPerformance.index, y=historicalPerformance['Close_True'], mode='lines', name='True'))

# Add Close_Predictions
# fig.add_trace(go.Scatter(x=historicalPerformance.index, y=historicalPerformance['Close_Predictions'], mode='lines', name='Predictions'))

# fig.add_trace(go.Scatter(x=historicalPerformance.index, y=historicalPerformance[target + '_Predictions'], mode='lines', name='Predictions'))

#make a shaded area for the prediction (width = rmse on each side)
fig.add_trace(go.Scatter(x=historicalPerformance.index, y=historicalPerformance['Close_Predictions'] * (1 + RMSE), fill=None, mode='lines', line_color='rgba(0,100,80,0.2)', name='Prediction_High'))
fig.add_trace(go.Scatter(x=historicalPerformance.index, y=historicalPerformance['Close_Predictions'] * (1 - RMSE), fill='tonexty', mode='lines', line_color='rgba(0,100,80,0.2)', name='Prediction_Band'))
fig.show()
```
