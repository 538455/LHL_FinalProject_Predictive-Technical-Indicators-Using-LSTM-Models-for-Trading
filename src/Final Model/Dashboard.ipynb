```python
# Scheduled Refresh of Predictions: 16:30 PM every weekday
from Functions import refreshPredictions
import schedule
import time

# set the schedule to run every weekday at 1630
schedule.every().monday.at("16:30").do(refreshPredictions)
schedule.every().tuesday.at("16:30").do(refreshPredictions)
schedule.every().wednesday.at("16:30").do(refreshPredictions)
schedule.every().thursday.at("16:30").do(refreshPredictions)
schedule.every().friday.at("16:30").do(refreshPredictions)

# run the schedule
while True:
    schedule.run_pending()
    time.sleep(3600) # 3600 seconds = 1 hour
    
```
