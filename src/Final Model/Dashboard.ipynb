```python
# Create Prev_Day_Accuracy column for close, open, high, low, Median_Price
SBB['Prev_Day_Accuracy_Close'] = SBB['Close_True'].shift(1) / SBB['Close_Predictions'].shift(1) *100 - 100
SBB['Prev_Day_Accuracy_Open'] = SBB['Open_True'].shift(1) / SBB['Open_Predictions'].shift(1) *100 - 100
SBB['Prev_Day_Accuracy_High'] = SBB['High_True'].shift(1) / SBB['High_Predictions'].shift(1) *100 - 100
SBB['Prev_Day_Accuracy_Low'] = SBB['Low_True'].shift(1) / SBB['Low_Predictions'].shift(1) *100 - 100
SBB['Prev_Day_Accuracy_Median_Price'] = SBB['Median_Price_True'].shift(1) / SBB['Median_Price_Predictions'].shift(1) *100 - 100

# Create 'PrevDayAccuracy' column that is the avg of the accuracy for close, high and median price of the previous day
SBB['PrevDayAccuracy3'] = (SBB['Prev_Day_Accuracy_Close'] + SBB['Prev_Day_Accuracy_High'] + SBB['Prev_Day_Accuracy_Median_Price']) / 3

# delete the accuracy columns
# SBB = SBB.drop(['Prev_Day_Accuracy_Open', 'Prev_Day_Accuracy_High', 'Prev_Day_Accuracy_Low', 'Prev_Day_Accuracy_Median_Price'], axis=1)

# Create PrevDayGotHigh column that is 1 if the previous day's predicted high was higher than the true high (shift 1)
SBB['PrevDayGotHigh'] = SBB['High_Predictions'] > SBB['High_True'].shift(1)

#same for close, low, open, median price
SBB['PrevDayGotClose'] = SBB['Close_Predictions'].shift(1) > SBB['Close_True'].shift(1)
# SBB['PrevDayGotLow'] = SBB['Low_Predictions'] < SBB['Low_True'].shift(1)
SBB['PrevDayGotOpen'] = SBB['Open_Predictions'].shift(1) > SBB['Open_True'].shift(1)
SBB['PrevDayGotMedianPrice'] = SBB['Median_Price_Predictions'].shift(1) > SBB['Median_Price_True'].shift(1)
```
